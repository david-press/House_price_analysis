# Data cleaning

```python
df = df.drop(columns = ["street","statezip","country"]) #to drop unnecessary colmns
```

```python
#to convert date from object ato datetime
df["date"] = pd.to_datetime(df["date"] , errors = 'coerce' , dayfirst = True)
#formats = '%Y-%m-%d'
```

```python
df.head()
```

# Feature engineering

## Gropby- pattern extraction

```python
#group price by city
df.groupby("city")["price"].agg(["mean","median","count"]).head()
```

```python
#Now here the main target is the 'price' column
df["price"].value_counts()
#to group the price 
df["price_group"] = pd.qcut(df["price"] , q = 3 , labels = ["Low_price","Medium_price","High_price"])

```

```python
df.head(4)
```

```python
df.groupby("price_group")["bedrooms"].agg(["mean","median","count"])
```

```python
df.groupby("price_group")["bathrooms"].agg(["mean","median","count"])
```

## correlation scanner - relationship scanner
(correlation measures linear relationships strength)

```python
#this is to find the top correlations (top 5 drivers of price)
corr = df.corr(numeric_only = True)  
top_cor = corr['price_log'].drop(['price','price_log']).abs().sort_values(ascending = False).head(7).index
```

```python
#to visualize the correlation
plt.figure(figsize = (10 ,8))
sns.heatmap(corr , cmap = "coolwarm")
plt.savefig('../images/correlation_heatmap.png', dpi=300, bbox_inches="tight")
plt.show()
```

# Distribution analysis

```python
df["price_log"] = np.log(df["price"])
```

```python
sns.histplot(df["price_log"] , kde =True)
#KDE -kernel distribution Estimate- a smooth curve distriibution
#this show there are few extremely expensive houses pulling the tail---- positive skew
```

# Top 5 drivers of house price with plots
(scatter plots for numerical variables)

```python
insights = {}
for col in top_cor:
    x=df[col]
    y=df["price"]
    #plot
    plt.figure()
    plt.scatter(x,y)
    plt.xlabel(col)
    plt.ylabel("price")
    plt.title(f"{col} vs price")
    plt.savefig(f'../images/{col}_vs_price.png', dpi=300, bbox_inches="tight")
    plt.show()
    
    #--Analysis
    corr_value = np.corrcoef(x,y)[0,1]
    if abs(corr_value) > 0.7:
        strength = "strong"
    elif abs(corr_value) > 0.4:
        strength = "moderate"
    else:
        strength = "weak"   
        
    #-detect skew
    skew = x.skew()
    skew_type = "Right-skewed" if skew > 0.5 else "Left-skewed" if skew < -0.5 else "Symetric"
    
    #outlier checks
    
    outliers = ((x - x.mean()).abs() > 3 * x.std()).sum()

    insights[col] = {
        "correlation" : round(corr_value,3),
        "strength" : strength,
        "distribution" : skew_type,
        "outliers_detected" : int(outliers)
    }
     
```

```python
for feature, info in insights.items():
    print(f"\nFeature : {feature}")
    print(f"\nCorrelation : {info['correlation']} {info['strength']}")
    print(f"Disribution : {info['distribution']}")
    print(f"Outliers : {info['outliers_detected']}")
    
```
